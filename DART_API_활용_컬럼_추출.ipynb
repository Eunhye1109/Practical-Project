```python
# 피클 임시저장용

import pickle

def save_pickle(obj, path: str):
    with open(path, "wb") as f:
        pickle.dump(obj, f)
    print(f"✅ 저장 완료 → {path}")

def load_pickle(path: str):
    with open(path, "rb") as f:
        obj = pickle.load(f)
    print(f"✅ 로드 완료 ← {path}")
    return obj
```

```python
import itertools
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def embed_and_compare(
    account_list: list[str],
    model_name: str = "snunlp/KR-SBERT-V40K-klueNLI-augSTS",
    top_k: int = 50
) -> pd.DataFrame:
    """
    account_list: 계정명 리스트
    model_name: 불러올 SentenceTransformer 모델
    top_k: 유사도 상위 N개 페어 반환
    """
    print(f"🔍 모델 로딩: {model_name}")
    model = SentenceTransformer(model_name)

    print(f"🔢 임베딩 중... ({len(account_list)}개)")
    embs = model.encode(account_list, convert_to_numpy=True)

    print(f"📐 유사도 계산...")
    sim = cosine_similarity(embs)
    pairs = [
        (account_list[i], account_list[j], sim[i, j])
        for i, j in itertools.combinations(range(len(account_list)), 2)
    ]
    top_pairs = sorted(pairs, key=lambda x: x[2], reverse=True)[:top_k]
    df = pd.DataFrame(top_pairs, columns=["Account1", "Account2", "Similarity"])
    return df
```

```python
def match_with_targets(
    all_accounts: list[str],
    target_keywords: list[str],
    model_name: str = "snunlp/KR-SBERT-V40K-klueNLI-augSTS",
    top_k: int = 5
) -> pd.DataFrame:
    print("📦 모델 로딩 중...")
    model = SentenceTransformer(model_name)

    print("🔢 전체 계정명 임베딩 중...")
    all_embs = model.encode(all_accounts, convert_to_numpy=True)

    print("🎯 타겟 키워드 임베딩 중...")
    target_embs = model.encode(target_keywords, convert_to_numpy=True)

    results = []

    for i, keyword in enumerate(target_keywords):
        sims = cosine_similarity([target_embs[i]], all_embs)[0]
        top_indices = sims.argsort()[::-1][:top_k]

        for idx in top_indices:
            results.append((keyword, all_accounts[idx], sims[idx]))

    df_result = pd.DataFrame(results, columns=["TargetKeyword", "MatchedAccount", "Similarity"])
    return df_result.sort_values(by=["TargetKeyword", "Similarity"], ascending=[True, False])
```

```python
top_500 = all_listed_corps.head(500)
corp_code_dict = dict(zip(top_500['corp_name'], top_500['corp_code']))
```

```python
account_rows = []

for corp_name in top_500['corp_name']:
    corp_code = corp_code_dict.get(corp_name)

    if not corp_code:
        print(f"⚠️ {corp_name} - corp_code 없음")
        continue

    found = False  # 성공 여부 플래그

    for fs_div in ['OFS', 'CFS']:  # 별도 → 연결 순으로 시도
        for reprt_code in ['11011', '11014', '11012']:  # 사업→반기→1분기
            try:

                data = dart_fss.api.finance.fnltt_singl_acnt_all(
                    corp_code=corp_code,
                    bsns_year='2021',
                    reprt_code=reprt_code,
                    fs_div=fs_div
                )

                df = pd.DataFrame(data['list'])

                if not df.empty:
                    df['corp_name'] = corp_name
                    account_rows.append(df[['corp_name', 'account_id', 'account_nm']])
                    print(f"✅ {corp_name} ({corp_code}) - {len(df)} rows from {fs_div}/{reprt_code}")
                    found = True
                    break  # 성공 시 더 이상 반복 안 함

            except Exception as e:
                continue  # 다음 조합 시도

        if found:
            break  # 성공했으면 fs_div 루프도 종료

    if not found:
        print(f"❌ {corp_name} ({corp_code}) - 데이터 없음 (모든 조합 실패)")
```

```python


# ✅ Step 0: 컬럼 정리
corp_columns = unique_accounts+ [
    ['매출액', '영업이익', '순이익'],
    ['매출', '영업수익', '당기순이익'],
    ['Revenue', 'OperatingProfit', 'NetIncome'],
    ['총매출액', '영업이득', 'Net Income'],
    ['매출총액', '영업이익', '순이익']
]

import re

def clean_name(name):
    # 한글, 알파벳, 숫자만 남기고 나머지 제거
    name = re.sub(r"[^가-힣a-zA-Z0-9]", "", name.lower())
    return name


all_columns = [clean_name(col) for corp in corp_columns for col in corp]
unique_columns = list(set(all_columns))


# ✅ Step 1: 컬럼명 임베딩 (의미 분석)
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embeddings = model.encode(unique_columns)


# ✅ Step 2: 클러스터링 (비슷한 컬럼끼리 그룹)
clustering = DBSCAN(eps=0.5, min_samples=2, metric='cosine').fit(embeddings)

cluster_map = {}
for idx, label in enumerate(clustering.labels_):
    if label == -1:
        continue
    cluster_map.setdefault(label, []).append(unique_columns[idx])

# ✅ Step 3: 전체 컬럼 빈도 수 세기
column_counter = Counter(all_columns)


# ✅ Step 4: 각 그룹에서 가장 많이 쓰인 컬럼 1개 대표로 뽑기
cluster_representatives = {}
for label, cols in cluster_map.items():
    most_common = sorted(cols, key=lambda x: column_counter[x], reverse=True)[0]
    cluster_representatives[most_common] = cols


# 매핑
account_mapping = {}
for rep, synonyms in cluster_representatives.items():
    for s in synonyms:
        account_mapping[s] = rep


# ✅ Step 5: 대표 컬럼 중에서 사용 빈도 높은 순으로 정렬
rep_counter = {
    rep: sum(column_counter[c] for c in synonyms)
    for rep, synonyms in cluster_representatives.items()
}


top_features = sorted(rep_counter.items(), key=lambda x: x[1], reverse=True)[:50]


# ✅ 결과 보기
import json
for feature, freq in top_features:
    print(f"[{feature}] (빈도: {freq}) → 유사 컬럼들: {json.dumps(cluster_representatives[feature], ensure_ascii=False)}")
```

```python
df_all['normalized'] = df_all['account_nm'].fillna('').apply(clean_name)
df_all['mapped'] = df_all['normalized'].map(account_mapping)
```

```python
import os
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch

# 1. 텍스트 파일 불러오기 및 항목명 추출
def extract_items_from_txt(folder_path):
    items = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            with open(os.path.join(folder_path, filename), 'r', encoding='cp949') as file:
                for line in file:
                    line = line.strip()
                    if line:  # 빈 줄 제외
                        items.append(line)
    return items

# 2. 유니크 항목명만 추출
def get_unique_items(items):
    return list(set(items))

# 3. 임베딩 모델 로드 (Ko-SBERT)
def load_model():
    model = SentenceTransformer('jhgan/ko-sbert-sts')
    return model.to('cuda')

# 4. 항목명 임베딩
def embed_items(model, items):
    return model.encode(items, convert_to_tensor=True, device='cuda')

# 5. 유사도 Top 50 계산
def compute_top50_similarities(items, embeddings, top_k=50):
    similarities = []
    for idx, name in enumerate(items):
        query_embedding = embeddings[idx]
        sim_scores = util.pytorch_cos_sim(query_embedding, embeddings)[0]
        top_results = torch.topk(sim_scores, k=top_k + 1)  # 자기 자신 포함
        similar_items = []
        for score, i in zip(top_results[0], top_results[1]):
            if i != idx:
                similar_items.append((items[i], float(score)))
            if len(similar_items) == top_k:
                break
        similarities.append({'항목명': name, '유사항목(Top50)': similar_items})
    return similarities

# 6. 결과 출력 (샘플)
def print_sample(similarities, sample_size=10):
    for s in similarities[:sample_size]:
        print(f"\n📌 {s['항목명']}")
        for sim_item, score in s['유사항목(Top50)'][:5]:
            print(f"    ↪ {sim_item} (유사도: {score:.4f})")

# --- 실행 ---
folder_path = '/content/drive/MyDrive/Colab Notebooks/실전프로젝트/DART DATA/손익계산서'  # txt 파일이 있는 폴더 경로로 수정
raw_items = extract_items_from_txt(folder_path)
unique_items = get_unique_items(raw_items)

model = load_model()
embeddings = embed_items(model, unique_items)

similarities = compute_top50_similarities(unique_items, embeddings)
print_sample(similarities, sample_size=20)
```

```python
import os
import pandas as pd

def load_txts_as_dataframe(folder_path, encoding='cp949', sep='\t'):
    """
    지정된 폴더 내의 모든 .txt 파일을 읽어서 하나의 DataFrame으로 통합
    - encoding: 기본 cp949 (필요시 'utf-8' 또는 'utf-8-sig'로 변경)
    - sep: 기본 '\t' (필요시 ',' 로 변경)
    """
    df_list = []

    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)
            try:
                df = pd.read_csv(file_path, encoding=encoding, sep=sep)
                df_list.append(df)
                print(f"✅ 로딩 성공: {filename}")
            except Exception as e:
                print(f"❌ 로딩 실패: {filename} → {e}")

    if df_list:
        combined_df = pd.concat(df_list, ignore_index=True)
        print(f"\n📦 총 {len(df_list)}개 파일 로딩 완료 / 행 개수: {len(combined_df)}")
        return combined_df
    else:
        raise ValueError("⚠️ 텍스트 파일을 읽지 못했습니다. 경로나 인코딩, 구분자를 확인하세요.")

# 🔍 사용 예시
folder_path = '/content/drive/MyDrive/Colab Notebooks/실전프로젝트/DART DATA/손익계산서'  # ← 경로 수정
df_all = load_txts_as_dataframe(folder_path)
```

```python
def compute_top_similarities(items, embeddings, top_k=10, max_print=100):
    similarities = []
    for idx, name in enumerate(items):
        query_embedding = embeddings[idx]
        sim_scores = util.pytorch_cos_sim(query_embedding, embeddings)[0]
        top_results = torch.topk(sim_scores, k=top_k + 1)  # 자기 자신 포함
        similar_items = []

        for score, i in zip(top_results[0], top_results[1]):
            if i != idx:
                similar_items.append((items[i], float(score)))
            if len(similar_items) == top_k:
                break

        similarities.append({'항목명': name, '유사항목(Top10)': similar_items})

    # 상위 100개만 출력
    for s in similarities[:max_print]:
        print(f"\n📌 {s['항목명']}")
        for sim_item, score in s['유사항목(Top10)']:
            print(f"    ↪ {sim_item} (유사도: {score:.4f})")

    return similarities

# --- 실행 ---
similarities = compute_top_similarities(unique_items, embeddings, top_k=10, max_print=100)
```
